```python
# Assigment for matrix
tensor=torch.IntTensor([[1,2,3],[4,5,6]])

tensor[1][2]=9
print (tensor)
print ("\n", 25*"*")


# Assigment for 3D tensor
tensor3d=torch.IntTensor([[[1,2,3],[2,3,4]],
                            [[5,6,7],[8,9,1]]])

tensor3d[1][1][0]=12  #1st bracket is for layers, 2nd is for rows, and 3rd is columns
print (tensor3d)
print ("\n", 25*"*")

tensor3d[1,1,1]=15
print (tensor3d)
print ("\n", 25*"*")

tensor3d[0,:,1]=torch.IntTensor([[0,0]])
print (tensor3d)
```

```python
#zeros
zeros1=torch.zeros(4)
print(zeros1)
print ("\n", 25*"*")

zeros2=torch.zeros(2,3)
print(zeros2)
print ("\n", 25*"*")

zeros3=torch.zeros(2,3,2)
print(zeros3)
```

```python
#ones
ones1=torch.ones(4)
print(ones1)
print ("\n", 25*"*")

ones2=torch.ones(2,3)
print(ones2)
print ("\n", 25*"*")

ones3= torch.ones(2,2)+5
print (ones3)
print ("\n", 25*"*")

ones4= torch.ones(2,2)*5
print (ones4)
print ("\n", 25*"*")

ones5=torch.ones(2,3,2)
print(ones5)
```

```python
#eye

eye1=torch.eye(3)  #its an square matrix
print(eye1) 
print ("\n", 25*"*")

eye2= torch.eye(3,2)
print (eye2)
```

```python
#rand

rand1=torch.rand(5) #ran(n) generate n random numbers between 0 and 1
print(rand1) 
print ("\n", 25*"*")

rand2=torch.rand(2,3) 
print(rand2)

r,c = rand2.size()  #matrix size, r is number of row, c is number of columns
num=rand2.numel()  #number of elements (r*c)
```

```python
#normal random

randn1=torch.randn(5) #generate gaussian destribution mean=0  std=1
print(randn1) 
print ("\n", 25*"*")

randn2=torch.randn(2,3) 
print(randn2)
```

```python
#randperm

randpr1=torch.randperm(5)   #returns a row vector containing a random permutation of the integers from 1 to n-1 without repeating elements
print(randpr1)
```

```python
#range

range1= torch.range(10,20)  #range (n,m) returns a row vector containing numbers between n and m (n=start, m=end) 
print (range1)
print ("\n", 25*"*")

range2= torch.range(10,20,2) #with step 2
print (range2)

```

```python
#linspace

lins1=torch.linspace(10,20,4)  #linspace (n,m,p) returns a row vector contrains p numbers between n and m 
print (lins1)
```

```python
#logspace

logs1=torch.logspace(2,3,4)  #logspace (n,m,p) returns a row vector contrains p numbers between 10^n and 10^m 
print (logs1)
print ("\n", 25*"*")

logs1_int=logs1.type(torch.IntTensor)  #change its type to integer
print (logs1_int)
```
